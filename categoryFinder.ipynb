```python
df = pd.read_csv('Daily Household Transactions.csv')
```

```python
print(df.head())
```

### pip install scikit-learn

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['Name'], df['Category'], test_size=0.2, random_state=42)
```

```python
# Create and train model
model = make_pipeline(
    TfidfVectorizer(),
    LogisticRegression(class_weight='balanced', max_iter=1000)
)
model.fit(x_train, y_train)
```

```python
vectorizer = model.named_steps['tfidfvectorizer']
print(vectorizer.get_feature_names_out())
```

```python
print(df['Category'].value_counts())
```

```python
print(model.predict(["UPI-GROWW INVEST TECH PV-GROWWNBT.ELEMENTS@HDFCBANK-HDFC0MERUPI-546764572896-PAID VIA ELEMENTS"]))
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```
